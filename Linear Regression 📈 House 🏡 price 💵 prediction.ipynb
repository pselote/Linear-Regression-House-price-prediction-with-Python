```python
## for 0-it gives fixed means, for 1- evey time mean changes
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1)
```

```python
print(np.mean(x_train))
print(np.mean(x_test))
print(np.mean(y_train))
print(np.mean(y_test))
```

```python
from sklearn.linear_model import LinearRegression
l=LinearRegression()
l.fit(x_train,y_train)
print("train test complete")
```

```python
y_pred=l.predict(x_test)
```

```python
d1=pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
d1
```

```python
from sklearn.metrics import mean_absolute_error
print("Mean Absolute error",mean_absolute_error(y_test,y_pred))
```

```python
sns.regplot(x=y_test,y=y_pred)
plt.title("Regression plot",size=12)
plt.ylabel("Price",size=15)
plt.xlabel("Avg. Area Number of Rooms",size=12)
plt.show()
```
